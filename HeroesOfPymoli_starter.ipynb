```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
Total_Players = len(purchase_data['SN'].value_counts())
purchase_data['Total Players'] = Total_Players
purchase_data.loc[:, ["Total Players"]].head(1)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Unique_items = len(purchase_data['Item ID'].value_counts())
avg_price = purchase_data['Price'].mean()
avg_price = np.round(avg_price, decimals = 2)
Purchase_Number = purchase_data['Purchase ID'].count()
Total_Revenue = purchase_data['Price'].sum()
Summary = {'Number of Unique Items':[Unique_items],
           'Average Price':[avg_price],
           'Number of Purchases':[Purchase_Number],
           'Total Revenue':[Total_Revenue]}
Summary_df = pd.DataFrame(data = Summary)
Summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

Gender_Percentage = ((purchase_data['Gender'].value_counts())/Total_Players * 100).round(2)
Male_Count = len(purchase_data.loc[purchase_data['Gender'] == 'Male'])
Female_Count = len(purchase_data.loc[purchase_data['Gender'] == 'Female'])
Other_Count = len(purchase_data.loc[purchase_data['Gender'] == 'Other / Non-Disclosed'])
Male_Percent = ((Male_Count/Total_Players)*100)
Female_Percent = ((Female_Count/Total_Players)*100)
Other_Percent = ((Other_Count/Total_Players)*100)


Demographic = {'Percentage of Players':[Male_Percent, Female_Percent, Other_Percent],
                  'Total Count':[Male_Count, Female_Count, Other_Count]}
Demographic_df = pd.DataFrame(data = Demographic)
Demographic_df['Percentage of Players'] = Demographic_df['Percentage of Players'].apply(np.round, decimals = 2)
Demographic_df.rename({0: 'Male', 1: 'Female', 2: 'Other / Non-Disclosed'})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


purchase_data['age group'] = pd.cut(purchase_data['Age'], age_bins, labels=group_names)
purchase_group = purchase_data.groupby('age group')
purchase_group["Age"].count()

age_below_10 =  len(purchase_data.loc[purchase_data['age group'] == '<10'])
age_10to14 = len(purchase_data.loc[purchase_data['age group'] == '10-14'])
age_15to19 = len(purchase_data.loc[purchase_data['age group'] == '15-19'])
age_20to24 = len(purchase_data.loc[purchase_data['age group'] == '20-24'])
age_25to29 = len(purchase_data.loc[purchase_data['age group'] == '25-29'])
age_30to34 = len(purchase_data.loc[purchase_data['age group'] == '30-34'])
age_35to39 = len(purchase_data.loc[purchase_data['age group'] == '35-39'])
age_above_40 = len(purchase_data.loc[purchase_data['age group'] == '40+'])

Percent_age_below_10 = (age_below_10 / Total_Players)*100
Percent_age_10to14 = (age_10to14 / Total_Players)*100
Percent_age_15to19 = (age_15to19 / Total_Players)*100
Percent_age_20to24 = (age_20to24  / Total_Players)*100
Percent_age_25to29 = (age_25to29 / Total_Players)*100
Percent_age_30to34 = (age_30to34 / Total_Players)*100
Percent_age_35to39 = (age_35to39 / Total_Players)*100
Percent_age_above_40 = (age_above_40 / Total_Players)*100

Age_Demographics = {'Percentage of Players':[Percent_age_below_10,Percent_age_10to14,Percent_age_15to19,Percent_age_20to24,
                                             Percent_age_25to29, Percent_age_30to34, Percent_age_35to39, Percent_age_above_40],
                    'Total Count':[age_below_10,age_10to14,age_15to19,age_20to24,age_25to29,age_30to34,age_35to39,age_above_40]}
Age_demographics_df = pd.DataFrame(data = Age_Demographics)
Age_demographics_df ['Percentage of Players'] = Age_demographics_df ['Percentage of Players'].apply(np.round, decimals = 2)
Age_demographics_df.rename({0:"<10", 1:"10-14", 2:"15-19", 3:"20-24", 4:"25-29", 5:"30-34", 6:"35-39", 7:"40+"})

```
